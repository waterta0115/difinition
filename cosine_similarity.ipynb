## コサイン類似度

```python
A = np.arange(5)                    # ここに値を入れる
B = np.array([1, 2, 3, 4, 100])     # ここに値を入れる
def cosine_similarity(A, B):

    naiseki = sum(A * B)

    def eauclidian_norm(arr):
        x = []
        for a in arr:
            x.append(a**2)
        return np.sqrt(sum(x))
    normA = eauclidian_norm(arr = A)
    normB = eauclidian_norm(arr = B)
    return naiseki/(normA * normB)
```

```python
cosine_similarity(A,B)
```
